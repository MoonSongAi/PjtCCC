```python
import cv2
import numpy as np

# 이미지 읽기
image = cv2.imread('D:/PjtCCC/3DModeling/aaa.jpg')  # 이미지 파일 경로를 지정하세요
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 블러링을 적용하여 노이즈 감소
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 엣지 검출
edges = cv2.Canny(blurred, 50, 150)

# 선 검출 (Hough Line Transform 사용)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)

# 수직선과 수평선만을 필터링하여 이미지에 그리기
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        # 기울기 계산
        if x1 == x2:  # 수직선
            cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        elif y1 == y2:  # 수평선
            cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        else:
            slope = (y2 - y1) / (x2 - x1)
            if abs(slope) < 0.1:  # 수평선 (기울기 0에 가까운)
                cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            elif abs(slope) > 10:  # 수직선 (기울기 무한대에 가까운)
                cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# 결과 이미지 크기 조정
scale_percent = 30  # 이미지 크기를 50%로 줄이기
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

# 결과 이미지 출력
cv2.imshow('Detected Lines', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def is_specific_color_line(image, x1, y1, x2, y2, color_range=((119, 120, 125), (228, 230, 230))):
    """선의 시작점과 끝점 사이의 모든 픽셀 색상이 특정 색상 범위 내에 있는지 확인"""
    line_thickness = 1
    mask = np.zeros_like(image)
    cv2.line(mask, (x1, y1), (x2, y2), (255, 255, 255), line_thickness)
    masked_image = cv2.bitwise_and(image, mask)
    
    # 색상 범위 설정
    lower_color, upper_color = color_range
    lower_color = np.array(lower_color, dtype=np.uint8)
    upper_color = np.array(upper_color, dtype=np.uint8)
    
    # 색상 범위 내에 있는지 확인
    color_mask = cv2.inRange(masked_image, lower_color, upper_color)
    return np.any(color_mask)

def get_line_color_range(image, x1, y1, x2, y2):
    """선의 시작점과 끝점 사이의 모든 픽셀 색상의 상한값과 하한값을 계산하여 반환"""
    line_thickness = 2
    mask = np.zeros_like(image)
    cv2.line(mask, (x1, y1), (x2, y2), (255, 255, 255), line_thickness)
    masked_image = cv2.bitwise_and(image, mask)
    
    # 마스크된 이미지에서 색상이 있는 부분만 추출
    line_pixels = masked_image[np.where((mask == 255).all(axis=2))]
    
    if line_pixels.size == 0:
        return None, None
    
    # RGB 값의 하한값과 상한값 계산
    lower_color = np.min(line_pixels, axis=0)
    upper_color = np.max(line_pixels, axis=0)
    
    return lower_color, upper_color


# 이미지 읽기
image = cv2.imread('D:/PjtCCC/3DModeling/aaa.jpg')  # 이미지 파일 경로를 지정하세요
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 블러링을 적용하여 노이즈 감소
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 엣지 검출
edges = cv2.Canny(blurred, 50, 150)

# 선 검출 (Hough Line Transform 사용)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)

# 선 길이가 200 이상이고 각도 오차를 포함한 수직선과 수평선만을 필터링하여 이미지에 그리기
angle_threshold = 5  # 각도 오차를 설정 (단위: 도)
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        length = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        if length >= 200 :
            angle = np.degrees(np.arctan2(y2 - y1, x2 - x1))
            is_box_line = False
            if ((abs(angle) < angle_threshold or abs(angle - 180) < angle_threshold)) and not is_specific_color_line(image, x1, y1, x2, y2):  # 수평선
                is_box_line = True
            elif ((abs(angle - 90) < angle_threshold or abs(angle + 90) < angle_threshold)) and not is_specific_color_line(image, x1, y1, x2, y2):  # 수직선
                is_box_line = True
            if is_box_line:
                lower_color, upper_color = get_line_color_range(image, x1, y1, x2, y2)
                if lower_color is not None and upper_color is not None:
                    print(f"Line from ({x1}, {y1}) to ({x2}, {y2}) - Lower color: {lower_color}, Upper color: {upper_color}")

                cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)


# 결과 이미지 크기 조정
scale_percent = 30  # 이미지 크기를 30%로 줄이기
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

# 결과 이미지 출력
cv2.imshow('Detected Lines', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
